```python
# ðŸŽ¯ The Detective's Tool - Data Folding

#What is "data folding"?

# Imagine taking a long strip of data and folding it every time 
# a planet completes one orbit. 
# All the transit dips will line up and become visible!

# Basically, if your guess about the orbital period is correct, 
# you would have a lot of data points in phase around 0.5. 

# Try it yourself!

# Run this cell, then experiment with different periods in the next cell.

def try_period(period_days):
    """
    Try folding the data at a specific period (in days).
    
    Parameters:
    period_days (float): The orbital period to test
    
    Example:
    try_period(1.5)  # Test a 1.5-day orbit
    """
    print(f"\nðŸ§ª Testing period: {period_days} days") # just a print line to know your guess. 
    mission.fold_at_period(period_days) # The funtion to fold is extensive and described in the source code. 
```
